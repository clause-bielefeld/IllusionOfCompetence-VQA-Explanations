### Chi-square Test of Independence: 

```python
cross_tab = pd.crosstab(answ_df['ANSW_CORRECT'], expl_df['ANSW_CORRECT'])

chi2_stat, p_val, _, _ = stats.chi2_contingency(cross_tab)

print("Chi-square Statistic:", chi2_stat)
print("P-value:", p_val)
```

- The chi-square test statistic was found to be χ² = 2.3617 with 2 degrees of freedom (df). The associated p-value was p = 0.6696; 
- p = 0.65 --> so there is no significant difference between the two experiments with respect to this evaluation criterion

## Statistical Tests 

- 5-point-likert scale = ordinal data 
- for ordinal data: non-paramateric tests ("Rang(summen)tests")
- e.g. Mann-Whitney U-test 

### Mann-Whitney U test 
The (Mann-Whitney) U-test asks whether the mean values (medians) of two independent measurement series signiﬁcantly differ from each other

Generally, the larger the U-statistic, the greater the difference between the distributions of the two samples.

```python
# Only process the black and white condition
bw_answ_df = answ_df[answ_df['COLOR_COND'] == 'bw']
bw_expl_df = expl_df[expl_df['COLOR_COND'] == 'bw']

columns_with_0s = ['COLORS', 'SHAPES', 'MATERIALS', 'GEN_SCENE', 'COMPETENT']
filtered_answ_df = bw_answ_df[~(bw_answ_df[columns_with_0s] == 0).any(axis=1)]
filtered_expl_df = bw_expl_df[~(bw_expl_df[columns_with_0s] == 0).any(axis=1)]

results_list = []  # Use a list to collect results

for column in columns_with_0s:
    # Mann-Whitney U test for the current column with explanation vs. without explanation
    data_answ = filtered_answ_df[column]
    data_expl = filtered_expl_df[column]
    u_statistic, p_value = stats.mannwhitneyu(data_answ, data_expl, alternative='two-sided')
    
    # Append the result as a dictionary to the list
    results_list.append({'Column': column, 'U-statistic': u_statistic, 'P-value': p_value})

# Convert the list of results into a DataFrame
results_df = pd.DataFrame(results_list)

# Display the results
results_df
```

## Automatic Measures  

```python
#### Automatic measures 

# Calculate BLEU scores for each row
def calculate_bleu(row):
    prediction = [str(row['model_explanation'])] 
    references = [row['human_explanations']]    
    bleu_score =  bleu_metric.compute(references=references, predictions = prediction)
    return bleu_score["precisions"][0]

# Apply function to each row of the DataFrame
experiment_data['bleu_scores'] = experiment_data.apply(calculate_bleu, axis=1)

## calculate ROUGE-L SCORE and add to the df
# Calculate ROUGE scores for each row
def calculate_rouge(row):
    prediction = [str(row['model_explanation'])]  
    references = [row['human_explanations']]    
    rouge_score =  rouge_metric.compute(references=references, predictions = prediction)
    return rouge_score["rougeL"]

# Apply the function to each row of the DataFrame
experiment_data['rouge_scores'] = experiment_data.apply(calculate_rouge, axis=1)

## calculate BERTSCORE and add to the df
# Calculate BERTscores for each row
def calculate_bertscore(row):
    prediction = [str(row['model_explanation'])] 
    references = [row['human_explanations']]    
    bert_score =  bertscore_metric.compute(references=references, predictions = prediction, lang="de")
    return bert_score["f1"][0]

# Apply the function to each row of the DataFrame
experiment_data['bert_scores'] = experiment_data.apply(calculate_bertscore, axis=1)
automatic_values = experiment_data
```
